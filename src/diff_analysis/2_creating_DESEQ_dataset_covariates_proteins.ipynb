# 1. Con Vs Prodromal

## 1.1 Creating dataset for con vs prodromal

```python
count_matrix = pd.read_csv('/12tb_dsk3/PD_classifier/Diff_exp_analysis/ppmi_orginal_paper_analysis/curated_dataset/count_matrix_con_prodromal.csv', index_col = 0)
count_matrix
```

```python
meta_data = pd.read_csv('/12tb_dsk3/PD_classifier/Diff_exp_analysis/ppmi_orginal_paper_analysis/curated_dataset/meta_data_con_prodromal.csv', index_col = 0)
meta_data
```

```python
meta_data.CLINICAL_EVENT.value_counts()
```

```python
import pandas as pd
import numpy as np



# Step 1: Convert raw counts to CPM
def counts_to_cpm(counts):
    # Calculate the total counts for each sample (sum of all genes in a sample)
    total_counts = counts.sum(axis=0)
    
    # Convert to CPM
    cpm = counts.div(total_counts, axis=1) * 1e6
    return cpm



# Step 2: Apply filtering based on CPM (similar to filterByExpr)
def filter_by_expr(cpm, min_samples, min_cpm=1):
    
    # Create a mask for filtering
    mask = np.zeros(cpm.shape[0], dtype=bool)

    # Iterate over genes and count how many samples meet the minimum CPM threshold
    for i, gene in enumerate(cpm.index):
        # For each gene, check the CPM values across samples
        gene_cpm = cpm.loc[gene]
        # Filter by CPM threshold
        filtered_cpm = gene_cpm[gene_cpm >= min_cpm]
        
        # Check if the gene has CPM >= min_cpm in at least 'min_samples' samples
        if len(filtered_cpm) >= min_samples:
            mask[i] = True

    # Filter the DataFrame using the mask
    filtered_cpm = cpm[mask]

    return filtered_cpm
```

```python
con_bl_data = count_matrix.loc[meta_data[meta_data.DIAGNOSIS == 'Control'].index]
pro_bl_data = count_matrix.loc[meta_data[meta_data.DIAGNOSIS == 'Prodromal'].index]
```

```python
countsSub = con_bl_data.T
cpm_data = counts_to_cpm(countsSub)
filtered_cpm = filter_by_expr(cpm_data, min_samples=countsSub.shape[1]//2, min_cpm=1)
con_bl_data_genes = filtered_cpm.index.tolist()
len(con_bl_data_genes)
```

```python
countsSub = pro_bl_data.T
cpm_data = counts_to_cpm(countsSub)
filtered_cpm = filter_by_expr(cpm_data, min_samples=countsSub.shape[1]//2, min_cpm=1)
pro_bl_data_genes = filtered_cpm.index.tolist()
len(pro_bl_data_genes)
```

```python
l  = set(pro_bl_data_genes).union(set(con_bl_data_genes))

con_bl_data = con_bl_data.loc[:,list(l)]
pro_bl_data = pro_bl_data.loc[:,list(l)]
```

```python
df = pd.concat([con_bl_data, pro_bl_data])
df
```

```python
meta_data = meta_data.loc[df.index]
```

```python
meta_data = meta_data.rename(columns = {'DIAGNOSIS': 'condition'})
```

```python
meta_data.columns
```

```python
df
```

```python
from Ensembl_converter import EnsemblConverter
```

```python
from Ensembl_converter import EnsemblConverter

# Create an instance of EnsemblConverter
converter = EnsemblConverter()

# Provide a list of Ensembl IDs
ensembl_ids = df.columns.tolist()

# Convert Ensembl IDs to gene symbols
result = converter.convert_ids(ensembl_ids)

# Print the resulting DataFrame
print(result)
```

```python
result
```

```python
dict_from_df = result.set_index('ENSG')['Symbol'].to_dict()
```

```python
df = df.rename(columns=dict_from_df )
```
