```r
library(EnhancedVolcano)
```

# Control Vs Prodromal

```r
data <- read.csv("con_vs_pro/BL//stat_results.csv")
head(data)
```

```r
rownames(data) <- data[, 1]
```

```r
colnames(data)[colnames(data) == "log2FoldChange"] <- "logFC"
head(data)
```

```r
rownames(data) <- data[, 1]
```

```r
svg("con_vs_pro/BL/CON_vs_PRO.svg", width = 15, height = 12)
EnhancedVolcano(
    title = 'Control vs Prodromal',
  data,
  lab =  rownames(data),
  x = 'logFC',
  y = 'padj',
  
  xlim = c(-0.6, 0.5),
  ylim = c(0, 1.4),
  pCutoff = 0.05,
  FCcutoff = 0.1,
  pointSize = 2.0,boxedLabels = TRUE
 
)


dev.off()
```

```r
EnhancedVolcano(
    title = 'Control vs Prodromal',
  data,
  lab =  rownames(data),
  x = 'logFC',
  y = 'padj',
  
  xlim = c(-0.6, 0.5),
  ylim = c(0, 1.4),
  pCutoff = 0.05,
  FCcutoff = 0.1,
  pointSize = 2.0,boxedLabels = TRUE
 
)
dev.off()
```

# 2. Control VS PD

```r
data <- read.csv("con_vs_pd/BL//stat_results.csv")
head(data)
```

```r
rownames(data) <- data[, 1]
```
