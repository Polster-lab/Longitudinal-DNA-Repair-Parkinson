```python
def extract_ensembl_ids_from_gtf(gtf_file):
    ensembl_gene_ids = []
    
    with open(gtf_file, "r") as gtf:
        for line in gtf:
            if line.startswith("#"):  # Skip comment lines
                continue
            fields = line.strip().split("\t")
            if fields[2] == "gene":
                attributes = fields[8]
                # Check if it's a protein-coding gene
                if 'gene_biotype "protein_coding"' in attributes:
                    # Extract the Ensembl gene ID
                    for attr in attributes.split(";"):
                        if "gene_id" in attr:
                            ensembl_gene_id = attr.split('"')[1]
                            ensembl_gene_ids.append(ensembl_gene_id)
                            break  # No need to check further attributes
    
    return ensembl_gene_ids

# Example usage
gtf_file = "../../Diff_exp_analysis/genomic_data/Homo_sapiens.GRCh38.104.gtf" # Replace with the path to your GTF file
ensembl_gene_ids = extract_ensembl_ids_from_gtf(gtf_file)
```

```python
keep_genes = set(data.columns).intersection(set(ensembl_gene_ids))
len(keep_genes)
```

```python
data = data[list(keep_genes)]
```
